```python
#When using machine learning algorithms we should always split our data into a training set and test set. 
#(If the number of experiments we are running is large, then we can should be dividing our data into 3 parts,
#namely — training set, development set and test set). In our case, we will also separate out some data 
#for manual cross checking.
```

```python
#DATA PREPARATION---> to dataset apoteleitai apo 767 astheneis opote, gia TRAIN tha baloume 650 deigmata, gia TEST 100 deigmata
#kai gia Cross Check tha exoyme 17.
```

```python
dfTrain = diabetesDF[:650]
dfTest = diabetesDF[650:750]
dfCheck = diabetesDF[750:]
```

```python
# Twra prepei na xexwrisoume ta labels me ta features(kai gia to training kai to test dataset)
#kai parallhla ta kanoyme kai NumPy arrays giati to modelo mas kanei process NumPy array Format
```

```python
trainLabel = np.asarray(dfTrain['Outcome'])
trainData = np.asarray(dfTrain.drop('Outcome',1))
testLabel = np.asarray(dfTest['Outcome'])
testData = np.asarray(dfTest.drop('Outcome',1))
```

```python
#san teleytaio step tha kanoyme normalize ta inputs mas 
#Machine Learning models often benefit substantially from input normalization. It also makes it easier for us to understand 
#the importance of each feature later, when we’ll be looking at the model weights. 
#We’ll normalize the data such that each variable has 0 mean and standard deviation of 1
```

```python
means = np.mean(trainData, axis=0)
stds = np.std(trainData, axis=0)
trainData = (trainData - means)/stds
testData = (testData - means)/stds
# np.mean(trainData, axis=0) => check that new means equal 0
# np.std(trainData, axis=0) => check that new stds equal 1
```

```python
diabetesCheck = LogisticRegression()
diabetesCheck.fit(trainData, trainLabel)
```

```python
accuracy = diabetesCheck.score(testData, testLabel)
print("accuracy = ", accuracy * 100, "%")
shap.initjs()
```

```python
explainer = shap.KernelExplainer(diabetesCheck.predict_proba, testData)
#shap_values = explainer.shap_values(trainData)
#shap.force_plot(explainer.expected_value[0], shap_values[0], trainData)
shap_values = explainer.shap_values(testData[0:])
shap.force_plot(explainer.expected_value[0], shap_values[0], testData[0:])
```

```python
ex = shap.KernelExplainer(diabetesCheck.predict, testData)
shap_values = ex.shap_values(testData[33,:])

fig=shap.force_plot(ex.expected_value, shap_values, testData[33,:], link='logit', show=False, matplotlib=True,
              feature_names=['Pregnancies', 'Glucose', 'Blood Pressure', 'Skin Thickness', 'Insulin',
                               'BMI', 'Diabetes Pedigree', 'Age'])
# fig = plt.gcf()
# fig.set_figheight(3)


# # fig.set_size_inches(45.5, 25.5)
```
